## 1. My lane (or freestyle) and why

Lane chosen: Lane 2: Refresh / Content Opportunity Scoring

Why: Recapturing traffic on pages that already rank but are actively decaying is often the fastest way to drive SEO ROI. Instead of blindly guessing which pages need updates, we can use historical search signals (like 90-day impressions and trend direction) to build a prioritization queue. This ensures human editors focus their limited time only on pages with enough volume to matter.


## 2. The question: decision, action, cost of a wrong call

The Research Question: Which decaying pages should the content review team prioritize for a refresh this week?
Unit of Analysis: A single page (content item).

The Decision: Determining the rank/priority order in which editors should review decaying pages.

The Action: The content team will rewrite meta descriptions, update outdated facts, or expand the content on the highest-ranked pages.

Cost of a wrong call:

False Positives: Wasting expensive human editor time on pages that don't need updates (e.g., normal seasonal dips).

False Negatives: Missing actual decaying pages, bleeding traffic and revenue to competitors while we do nothing.

Why ML helps: A hand-written rule (like the baseline) is too rigid. ML can weight multiple weak signals (CTR gaps, age, engagement) to find decaying pages that fixed rules miss, maximizing the ROI of the editors' time.

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd

# Use the raw GitHub URL to bypass Colab's folder structure issues
url = "https://raw.githubusercontent.com/MananAslamDev/ML-Stuff/main/data/raw/content_refresh_anonymized.csv"
df = pd.read_csv(url)

# Calculate our numbers
total_pages = len(df)
actionable = df[(df['trend_direction'] == 'down') & (df['impressions_90d'] >= 500)]
actionable_count = len(actionable)
traffic_at_risk = actionable['impressions_90d'].sum()

print(f"1. Total pages in the starter dataset: {total_pages:,}")
print(f"2. Actionable declining pages (>= 500 impressions): {actionable_count:,}")
print(f"3. 90-day impressions currently at risk on these pages: {traffic_at_risk:,.0f}")
print(f"\nConclusion: This targeted group represents {(actionable_count/total_pages)*100:.1f}% of the inventory, making it a highly focused and high-value review queue for the next 7 weeks.")
```

## 4. Careful words: what I can and can't claim

What I can claim: I can claim which pages exhibit the strongest historical signals of visibility decay compared to their peers based on observable metrics.

What I cannot claim: I cannot claim that refreshing the page will cause a guaranteed traffic recovery. This data is purely observational, not an A/B test. I am building a decision-support queue, not predicting causal outcomes. I also cannot claim this proves anything about Google's core algorithm.
